```python
import os 
import glob 
import osl_imitation_env  # déclenche le register()


def evaluate_imitation_quality(model, env_id='myoOslImitation-v0', n_episodes=10, seed=999):
    env = gym.make(env_id, seed=seed)
    all_diffs = []

    for ep in range(n_episodes):
        obs, info = env.reset()
        terminated, truncated = False, False
        while not (terminated or truncated):
            action, _ = model.predict(obs, deterministic=True)
            obs, reward, terminated, truncated, info = env.step(action)
            q_diff = env.unwrapped._get_qpos_diff_array()
            all_diffs.append(np.abs(q_diff))

    all_diffs = np.array(all_diffs)
    env.close()
    return {
        "mean_abs_diff_per_joint": all_diffs.mean(axis=0),
        "overall_mean_abs_diff": all_diffs.mean(),
    }

def list_available_runs(eval_logs_dir="./eval_logs"):
    """Liste tous les runs pour lesquels un fichier evaluations.npz existe."""
    runs = []
    for path in glob.glob(f"{eval_logs_dir}/*/evaluations.npz"):
        run_name = os.path.basename(os.path.dirname(path))
        runs.append(run_name)
    return sorted(runs)


def find_latest_checkpoint(run_name, checkpoints_dir="./checkpoints"):
    """Trouve le checkpoint le plus avancé (le plus de steps) pour un run donné."""
    pattern = f"{checkpoints_dir}/{run_name}/{run_name}_*_steps.zip"
    candidates = glob.glob(pattern)
    if not candidates:
        # repli sur best_model ou _final
        for fallback in [f"{checkpoints_dir}/{run_name}/best_model.zip",
                          f"{checkpoints_dir}/{run_name}_final.zip"]:
            if os.path.exists(fallback):
                return fallback
        return None
    # trie par nombre de steps extrait du nom de fichier
    candidates.sort(key=lambda p: int(p.split("_")[-2]))
    return candidates[-1]
```

```python
path_init0 = find_latest_checkpoint("imitation_init0_sparse1_solved10_qpos40")
model_init0 = PPO.load(path_init0)
results_init0 = evaluate_imitation_quality(model_init0, n_episodes=10)

print("=== imitation_init0 ===")
print(f"Écart moyen absolu global : {results_init0['overall_mean_abs_diff']:.4f} rad")
print(f"Écart moyen par joint : {results_init0['mean_abs_diff_per_joint']}")

path_imit_50M  = find_latest_checkpoint("imitation_sparse1_solved10_qpos40")
model_imit_50M  = PPO.load(path_imit_50M)
results_imit_50M  = evaluate_imitation_quality(model_imit_50M, n_episodes=10)


print("\n=== imitation_50M (ancien, bugué) ===")
print(f"Écart moyen absolu global : {results_imit_50M['overall_mean_abs_diff']:.4f} rad")
print(f"Écart moyen par joint : {results_imit_50M['mean_abs_diff_per_joint']}")
```

 ## Sanity check data

```python
from gymnasium.envs.registration import register
from myosuite.utils import gym
import osl_imitation_env
import numpy as np


def check_reference_data(env_id="myoOslImitation-v0"):
    print(f"=== INITIALISATION DE L'ENVIRONNEMENT : {env_id} ===")
    env = gym.make(env_id)
    env.reset()
    
    # Récupération de l'instance de votre classe (unwrap du wrapper gym)
    raw_env = env.unwrapped
    
    print("\n--- 1. Vérification des dimensions et en-têtes ---")
    print(f"Forme des données CSV (INIT_DATA) : {raw_env.INIT_DATA.shape}")
    print(f"Nombre de joints biologiques suivis : {len(raw_env.biological_jnt)}")
    print(f"Pas de temps de la simulation (dt) : {raw_env.dt} secondes")
    
    # Vérification de la correspondance des en-têtes
    missing_jnts = [j for j in raw_env.biological_jnt if j not in raw_env.gait_cycle_headers]
    if missing_jnts:
        print(f"⚠️ ATTENTION : Les joints suivants ne sont pas dans le CSV : {missing_jnts}")
    else:
        print("✅ Tous les joints biologiques sont présents dans les en-têtes du CSV.")

    print("\n--- 2. Test d'alignement Cinématique (Pas à Pas) ---")
    # On va lire les 5 premières lignes du CSV pour voir si MuJoCo accepte les poses
    for step_idx in range(5):
        raw_env._set_pose_from_reference(step_idx)
        
        # Extraction des variables d'observation après mise à jour de la pose
        obs_dict = raw_env.obs_dict
        root_vel = obs_dict.get("model_root_vel", np.zeros(3)).squeeze()
        socket_forces = obs_dict.get("socket_forces", np.zeros(3)).squeeze()
        
        # Calcul de vos fonctions custom de récompense à blanc
        q_diff = raw_env._get_qpos_diff_array()
        r_imitation = np.mean(np.exp(-8 * np.square(q_diff)))
        r_speed = raw_env._get_forward_bounded_reward()
        
        print(f"\n[Ligne CSV {step_idx}]")
        print(f"  -> Écart moyen absolu des joints (q_diff) : {np.mean(np.abs(q_diff)):.4f}")
        print(f"  -> Score d'imitation calculé (sans dt) : {r_imitation:.4f}")
        print(f"  -> Vitesse racine (MuJoCo) : {root_vel} (Axe Y de progression = index 1)")
        print(f"  -> Forces au niveau du manchon (Socket) : {socket_forces}")
        print(f"  -> Reward de vitesse calculé : {r_speed:.6f}")

    env.close()

check_reference_data()
```

```python
from gymnasium.envs.registration import register
from myosuite.utils import gym
import osl_imitation_env
import numpy as np

import time

def play_reference_trajectory(env_id="myoOslImitation-v0", loops=2):
    """
    Joue la trajectoire de référence en boucle dans le visualiseur passif de MuJoCo.
    Permet de valider visuellement la qualité du dataset de marche.
    """
    print("=== LANCEMENT DE LA VISUALISATION CINÉMATIQUE ===")
    # On force le mode de rendu humain pour ouvrir la fenêtre MuJoCo
    env = gym.make(env_id, render_mode="human")
    env.reset()
    raw_env = env.unwrapped
    
    total_rows = raw_env.INIT_DATA.shape[0]
    print(f"Lecture de {total_rows} frames de marche...")

    for loop in range(loops):
        print(f"Début du cycle de marche de référence - Boucle {loop+1}/{loops}")
        
        for idx in range(total_rows):
            # 1. On force la pose de la ligne courante du CSV
            raw_env._set_pose_from_reference(idx)
            
            # 2. On appelle render() pour mettre à jour la fenêtre graphique
            env.render()
            
            # 3. Petite pause pour caler la visualisation sur le temps réel (ex: 50-100Hz)
            time.sleep(0.02) 
            
            # Optionnel : afficher périodiquement les infos dans la console
            if idx % 20 == 0:
                print(f"  Frame {idx}/{total_rows} - Index d'imitation actuel : {raw_env._imitation_index}")
                
    env.close()
    print("Visualisation terminée.")

play_reference_trajectory()
```
